```python
def replacer(text: str) -> str:
  for word in forbiddenMarks:
    text = text.replace(word, ' ')
  return text

# abc -- 97 98 99 -> 97+98+99 => ~300
```

```python
def cleanMsg(msg: str) -> str:
  msg = replacer(msg)

  msg = msg.lower()
  words = msg.split()
  newMsg = []
  for i, word in enumerate(words):
    if not word in forbiddenWords:
      newMsg.append(word)

  print(' '.join(newMsg))
  return ' '.join(newMsg)
```

```python
# Очистить сообщения от каши(Привет, Спасибо, знаки препинания)
file_path = "Data.txt"
new_file = "newData.txt"

with open(file_path, 'r', encoding = 'utf-8') as srcFile:
  with open(new_file, 'w', encoding = 'utf-8') as newFile:
    lines = srcFile.readlines()
    for line in lines:

      newFile.write(cleanMsg(line) + "\n")
```

```python
import re
from pymorphy2 import MorphAnalyzer
import nltk
```

```python
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
patterns = "[A-Za-z0-9!#$%&'()*+,./:;<=>?@[\]^_`{|}~—\"\-]+"
stopwords_ru = stopwords.words("russian")
morph = MorphAnalyzer()
```

```python
def TLS(line):
  # line = row.values[0]
  line = re.sub(patterns, ' ', line)
  tokens = []
  for token in line.split():
    if token and token not in stopwords_ru:
      token = token.strip()
      token = morph.normal_forms(token)[0]

      tokens.append(token)

  # print(tokens)
  if len(tokens) > 2:
    return tokens
  return None

```

```python
columns = []
for i in range(len(data)):
  columns.append(TLS(data.loc[i+1,'message']))
```

```python
df = pd.DataFrame(columns = ['tokens'])
df.head()
```

```python
df['tokens'] = columns
display(df.dtypes)
```
